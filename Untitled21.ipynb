4. Statistical Analysis

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/GSS7402.csv"
df = pd.read_csv(url)

# Drop missing values for the statistical analysis
df_clean = df.dropna()

### 1. Descriptive Statistics (Numerical Columns)
print("Descriptive Statistics (Numerical Columns):\n", df_clean.describe())

### 2. Correlation Matrix for Numerical Variables
# Selecting only numeric columns for correlation
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
correlation_matrix = df_clean[numeric_cols].corr()
print("Correlation Matrix:\n", correlation_matrix)

# Visualizing Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

### 3. T-Test (Comparing Means of Two Groups)
# Hypothesis: Mean age is different for those who are immigrants and those who are not.

# Convert categorical 'immigrant' column to binary 1/0 for statistical analysis
df_clean.loc[:, 'immigrant_binary'] = df_clean['immigrant'].apply(lambda x: 1 if x == 'yes' else 0)

# Performing T-test
immigrant_group = df_clean[df_clean['immigrant_binary'] == 1]['age']
non_immigrant_group = df_clean[df_clean['immigrant_binary'] == 0]['age']
t_stat, p_val = stats.ttest_ind(immigrant_group, non_immigrant_group, nan_policy='omit')

print(f"T-Test: Comparing Age of Immigrants vs Non-Immigrants\nT-Statistic: {t_stat}, P-Value: {p_val}")

### 4. Chi-Square Test (Categorical Variables)
# Hypothesis: There is an association between 'ethnicity' and 'lowincome16'

# Cross-tabulation of the two categorical variables
contingency_table = pd.crosstab(df_clean['ethnicity'], df_clean['lowincome16'])

# Performing the chi-square test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square Test: Ethnicity vs Low Income at 16\nChi2 Statistic: {chi2_stat}, P-Value: {p_val}, Degrees of Freedom: {dof}")

### 5. ANOVA (Analysis of Variance)
# Hypothesis: There is a difference in the mean number of siblings across different ethnicities.

# Grouping data by ethnicity for ANOVA
ethnicity_groups = {
    'cauc': df_clean[df_clean['ethnicity'] == 'cauc']['siblings'].dropna(),
    'black': df_clean[df_clean['ethnicity'] == 'black']['siblings'].dropna(),
    'other': df_clean[df_clean['ethnicity'] == 'other']['siblings'].dropna()
}

# Check sample sizes before performing ANOVA
sample_sizes = {key: len(value) for key, value in ethnicity_groups.items()}
print(f"Sample sizes for ANOVA:\n{sample_sizes}")

# Perform ANOVA only if all groups have enough samples
if all(size > 1 for size in sample_sizes.values()):
    anova_result = stats.f_oneway(*ethnicity_groups.values())
    print(f"ANOVA: Siblings Across Ethnicities\nF-Statistic: {anova_result.statistic}, P-Value: {anova_result.pvalue}")
else:
    print("Insufficient sample size for ANOVA.")

### 6. Normality Test (Shapiro-Wilk Test)
# Hypothesis: Age follows a normal distribution.

shapiro_stat, shapiro_p_val = stats.shapiro(df_clean['age'])
print(f"Shapiro-Wilk Test for Age Normality\nShapiro Statistic: {shapiro_stat}, P-Value: {shapiro_p_val}")

# Visualizing the Age Distribution
sns.histplot(df_clean['age'], kde=True)
plt.title("Age Distribution")
plt.show()

### 7. Homogeneity of Variance (Levenes Test)
# Hypothesis: Variance in 'siblings' is equal across ethnicities.

# Perform Levene's test only if there are sufficient samples
if all(size > 1 for size in sample_sizes.values()):
    levene_stat, levene_p_val = stats.levene(*ethnicity_groups.values())
    print(f"Levenes Test for Homogeneity of Variance in Siblings Across Ethnicities\nLevene Statistic: {levene_stat}, P-Value: {levene_p_val}")
else:
    print("Insufficient sample size for Levene's test.")

### 8. Correlation between Age and Education (Spearman's Rank Correlation)
# Hypothesis: There is a monotonic relationship between age and education.

spearman_corr, spearman_p_val = stats.spearmanr(df_clean['age'], df_clean['education'])
print(f"Spearman's Rank Correlation: Age vs Education\nSpearman Correlation: {spearman_corr}, P-Value: {spearman_p_val}")

# Visualizing Age vs Education Correlation
sns.scatterplot(x='age', y='education', data=df_clean)
plt.title("Age vs Education Scatterplot")
plt.show()
```

8. Time Series Analysis (optional)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/GSS7402.csv"
df = pd.read_csv(url)

# Basic data preprocessing: drop rows with missing values in the 'year' column
df_clean = df.dropna(subset=['year'])

# Convert 'year' to int type for analysis
df_clean['year'] = df_clean['year'].astype(int)

# 1. Mean Age over Years
mean_age_by_year = df_clean.groupby('year')['age'].mean().reset_index()

# Plotting the mean age over years
plt.figure(figsize=(12, 6))
plt.plot(mean_age_by_year['year'], mean_age_by_year['age'], marker='o', color='blue')
plt.title('Mean Age Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Age')
plt.xticks(mean_age_by_year['year'])
plt.grid()
plt.show()

# 2. Mean Education Level over Years
mean_education_by_year = df_clean.groupby('year')['education'].mean().reset_index()

# Plotting the mean education over years
plt.figure(figsize=(12, 6))
plt.plot(mean_education_by_year['year'], mean_education_by_year['education'], marker='o', color='orange')
plt.title('Mean Education Level Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Education Level')
plt.xticks(mean_education_by_year['year'])
plt.grid()
plt.show()

# 3. Count of Immigrants over Years
immigrant_count_by_year = df_clean[df_clean['immigrant'] == 'yes'].groupby('year').size().reset_index(name='count')

# Plotting the count of immigrants over years
plt.figure(figsize=(12, 6))
plt.plot(immigrant_count_by_year['year'], immigrant_count_by_year['count'], marker='o', color='green')
plt.title('Count of Immigrants Over Years')
plt.xlabel('Year')
plt.ylabel('Count of Immigrants')
plt.xticks(immigrant_count_by_year['year'])
plt.grid()
plt.show()

# 4. Count of Individuals by Siblings over Years
siblings_count_by_year = df_clean.groupby(['year', 'siblings']).size().unstack(fill_value=0)

# Plotting the count of individuals by siblings over years
siblings_count_by_year.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Count of Individuals by Number of Siblings Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Number of Siblings')
plt.grid()
plt.show()
```
